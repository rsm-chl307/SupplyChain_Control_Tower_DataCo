```python
import pandas as pd
```

```python
df = pd.read_csv(
    "../data/raw/DataCoSupplyChainDataset.csv",
    encoding="latin1"
)

print(df.shape)

df.info()
```

```python
summary = pd.DataFrame({
    "dtype": df.dtypes.astype(str),
    "null_pct": round(df.isnull().mean()*100,2)
})

summary.sort_values(
    "null_pct",
    ascending=False
)
```
