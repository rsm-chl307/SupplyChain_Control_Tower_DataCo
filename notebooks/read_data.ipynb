```python
import kagglehub
import shutil
import os

path = kagglehub.dataset_download(
    "shashwatwork/dataco-smart-supply-chain-for-big-data-analysis"
)

print(path)

os.makedirs("data/raw", exist_ok=True)

for file in os.listdir(path):
    shutil.copy(
        os.path.join(path, file),
        "data/raw"
    )
```
